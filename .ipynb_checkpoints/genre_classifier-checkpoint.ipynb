```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(''"Webtoon Dataset.csv")
```

# Drop unnecessary columns from the DataFrame to clean the dataset.
- The following columns are being removed:
-'id' - Unique identifier for each record.
- 'Likes' - Number of likes, which may not be relevant for classification.
- 'Rating' - Rating value, potentially redundant or not useful for genre classification.
- 'Subscribers' - Number of subscribers, not needed for the analysis.
- 'Update' - Date of the last update, not relevant for genre classification.
- 'Reading Link' - Link to the webtoon, not needed for the model.

```python
data = data.drop(['id','Likes','Rating','Subscribers','Update','Reading Link', 'Writer'], axis=1)
```

```python
data.head()
```

# Exploring Data

```python
data.size
```

```python
print(len(data['Genre'].unique()))
print('--------------------------------------')
print(data['Genre'].unique())
```

```python
sns.countplot(data['Genre'])
```

```python
data['Genre'].value_counts().nsmallest(5)
```

```python
# List of airlines to exclude
excluded_genres = ['Historical', 'Heartwarming', 'Mystery', 'Informative','Sports']

# Filter out rows where 'airline' is in the excluded list
data = data[~data['Genre'].isin(excluded_genres)]
```

```python
data['Summary'] = data['Summary'].str.lower()
data['Name'] = data['Name'].str.lower()
```

# Checking for Null Values

```python
data.isna().sum()
```

# Vectorizing Textual Data

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data['text'] = data["Name"] + " " + data['Summary']
```

```python
vectorizer_name = TfidfVectorizer(stop_words='english')
X_text = vectorizer_name.fit_transform(data['text'])
```

# Train Test Split

```python
y = data['Genre']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_text, y, test_size=0.33, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42, k_neighbors=1)  
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
```

```python
sns.countplot(y_resampled)
```

# Model Selection

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay,accuracy_score
```

```python
def run_model(model): 
    # Fit Model
    model.fit(X_resampled,y_resampled)
    # Get Metrics
    preds = model.predict(X_test)
    print("accuracy " , accuracy_score(y_test,preds)*100)
    print(classification_report(y_test,preds))
    ConfusionMatrixDisplay.from_predictions(y_test,preds)
```

```python
logistic = LogisticRegression()
run_model(logistic)
```

```python
svm = SVC()
param_grid = {'C':[0.01,0.1,1,10],'kernel':['linear','rbf']}
grid_SVR = GridSearchCV(svm,param_grid)
```

```python
run_model(grid_SVR)
```

```python
n_estimators=[1,64,100,128]
max_features= [2,3,4]
bootstrap = [True,False]

param_grid = {'n_estimators':n_estimators,
             'max_features':max_features,
             'bootstrap':bootstrap} 
```

```python
rf = RandomForestClassifier()
grid_RF = GridSearchCV(rf,param_grid)
```

```python
run_model(grid_RF)
```
